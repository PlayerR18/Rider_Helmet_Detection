```python
import subprocess
import tkinter as tk
import os
import threading
import time
from tkinter import *
from tkinter import ttk
from PIL import Image, ImageTk
```

```python
# Set up the GUI
root = tk.Tk()
root.geometry('800x700')
root.title("Helmet Detector")
root.configure(bg='#000000')

# Center the GUI
win_width = root.winfo_reqwidth()
win_height = root.winfo_reqheight()
pos_x = int(root.winfo_screenwidth() / 2 - win_width / 2)
pos_y = int(root.winfo_screenheight() / 2 - win_height / 2)
root.geometry("+{}+{}".format(pos_x, pos_y))

# Define a custom style for the buttons
style = ttk.Style()
style.theme_use('default')
style.configure("Custom.TButton", background='#FFFFFF', foreground='#000000', font=("Arial", 20), padding=(10, 5))
style.map("Custom.TButton",
          background=[('active', '#212121'), ('pressed', '#424242')],
          foreground=[('active', '#FFFFFF'), ('pressed', '#FFFFFF')])

# Define the functions for running the notebooks and updating the progress bars
def run_ipynb():
    status_bar.config(text="Running Real Time System")
    process=subprocess.Popen(["python", "realtime.py"])
    try:
        for i in range(150):
            time.sleep(0.075)  # simulate some work
            pb['value'] = i
            text.config(text=f"{i}%")
            root.update_idletasks()
    except Exception as e:
        print(f"Error running Jupyter program: {e}")
    finally:
        pb.stop()
        text.config(text="Done")
        status_bar.config(text="Ready")

def run_video():
    status_bar.config(text="Running Video Detection System")
    process=subprocess.Popen(["python", "Video.py"])
    try:
        for i in range(150):
            time.sleep(0.075)  # simulate some work
            pb['value'] = i
            text.config(text=f"{i}%")
            root.update_idletasks()
    except Exception as e:
        print(f"Error running Jupyter program: {e}")
    finally:
        pb.stop()
        text.config(text="Done")
        status_bar.config(text="Ready")

def run_image():
    status_bar.config(text="Running Image Helmet Detection System")
    process=subprocess.Popen(["python", "C:/Users/pruth/Downloads/Final/image1.py"])
    try:
        for i in range(150):
            time.sleep(0.075)  # simulate some work
            pb['value'] = i
            text.config(text=f"{i}%")
            root.update_idletasks()
    except Exception as e:
        print(f"Error running Jupyter program: {e}")
    finally:
        pb.stop()
        text.config(text="Done")
        status_bar.config(text="Ready")

# Define the widgets for the GUI
title_label = Label(root, text="HELMET DETECTION SYSTEM", font=("Arial", 30), fg="white", bg="#000000")
title_label.place(relx=0.5, rely=0.1, anchor=CENTER)

image = Image.open("hell.png")
resize_image = image.resize((300, 300))
img = ImageTk.PhotoImage(resize_image)
image_label = Label(image=img, bg="#000000")
image_label.place(relx=0.5, rely=0.35, anchor=CENTER)

video_button = ttk.Button(root, text="Video Detection", style="Custom.TButton", command=run_video)
video_button.place(relx=0.3, rely=0.75, anchor=CENTER)

realtime_button = ttk.Button(root, text="Real Time Detection", style="Custom.TButton", command=run_ipynb)
realtime_button.place(relx=0.5, rely=0.75, anchor=CENTER)

image_button = ttk.Button(root, text="Image Detection", style="Custom.TButton", command=run_image)
image_button.place(relx=0.7, rely=0.75, anchor=CENTER)

pb = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
pb.place(relx=0.5, rely=0.9, anchor=CENTER)

text = Label(root, text="", font=("Arial", 20), fg="white", bg="#000000")
text.place(relx=0.5, rely=0.85, anchor=CENTER)

status_bar = Label(root, text="Ready", font=("Arial", 12), fg="white", bg="#000000", bd=1, relief=SUNKEN, anchor=W)
status_bar.pack(side=BOTTOM, fill=X)
```

```python
root.mainloop()
```
